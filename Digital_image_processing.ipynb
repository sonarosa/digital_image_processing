```python
import cv2
import matplotlib.pyplot as plt

# Reading the RGB image
skI = cv2.imread("dog.jpg")

# Convert the image from BGR to RGB (OpenCV loads images in BGR format)
skI_rgb = cv2.cvtColor(skI, cv2.COLOR_BGR2RGB)

# Displaying the original image
plt.subplot(1, 2, 1)
plt.imshow(skI_rgb)
plt.title("Original Image")

# Levels of the 8-bit image
L = 2 ** 8

# Finding the negative of the image
neg = (L - 1) - skI_rgb

# Displaying the negative image
plt.subplot(1, 2, 2)
plt.imshow(neg)
plt.title("Negative Image")

# Show the plot
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

def image_negative(image):
    """Apply negative transformation to a grayscale image."""
    L = 2 ** 8  # Levels of the 8-bit image
    return (L - 1) - image

# Load a grayscale image
grayscale_image = cv2.imread("dog.jpg", cv2.IMREAD_GRAYSCALE)

# Apply the image negative transformation
negative_image = image_negative(grayscale_image)

# Display the original and negative images
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(grayscale_image, cmap='gray')
plt.title("Original Grayscale Image")

plt.subplot(1, 2, 2)
plt.imshow(negative_image, cmap='gray')
plt.title("Negative Image")

plt.show()
```

```python
# Load low contrast and high contrast images
low_contrast_image = cv2.imread("dog.jpg", cv2.IMREAD_GRAYSCALE)
high_contrast_image = cv2.imread("dog.jpg", cv2.IMREAD_GRAYSCALE)

# Apply the image negative transformation
low_contrast_negative = image_negative(low_contrast_image)
high_contrast_negative = image_negative(high_contrast_image)

# Display the results
plt.figure(figsize=(10, 10))

plt.subplot(2, 2, 1)
plt.imshow(low_contrast_image, cmap='gray')
plt.title("Low Contrast Image")

plt.subplot(2, 2, 2)
plt.imshow(low_contrast_negative, cmap='gray')
plt.title("Negative of Low Contrast Image")

plt.subplot(2, 2, 3)
plt.imshow(high_contrast_image, cmap='gray')
plt.title("High Contrast Image")

plt.subplot(2, 2, 4)
plt.imshow(high_contrast_negative, cmap='gray')
plt.title("Negative of High Contrast Image")

plt.show()
```

```python
def plot_histogram(image, title):
    """Plot the histogram of a grayscale image."""
    plt.hist(image.ravel(), 256, [0, 256])
    plt.title(title)
    plt.show()

# Plot the histogram of the original grayscale image
plot_histogram(grayscale_image, "Histogram of Original Grayscale Image")

# Plot the histogram of the negative image
plot_histogram(negative_image, "Histogram of Negative Image")
```
